**(2) 네이버 api 활용**

네이버 api는 다양한 기능을 지원하고 있다. 네이버 아이디로 로그인이나 검색, 단축url이나 데이터랩 등을 지원하는데 검색어 트렌드 기능은 데이터랩의 일부라고 할 수 있다.<br>
<br>
그 중 검색을 이용해 뉴스를 검색해보려 했다.<br>
```python
# 네이버에서 제공하는 예시 코드
import os
import sys
import urllib.request
client_id = "YOUR_CLIENT_ID"
client_secret = "YOUR_CLIENT_SECRET"
encText = urllib.parse.quote("검색할 단어")
url = "https://openapi.naver.com/v1/search/news?query=" + encText # json 결과
# url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # xml 결과
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
response = urllib.request.urlopen(request)
rescode = response.getcode()
if(rescode==200):
    response_body = response.read()
    print(response_body.decode('utf-8'))
else:
    print("Error Code:" + rescode)
```
그러나 위의 코드를 실행하여 결과를 찾으려하니 기간에 대한 설정도 불가능했고 검색 결과도 최신순으로부터 1099건만이 검색이 불가능하다는 문제점이 있었다. 그리고 하루에 25000번이라는 최대 제약이 걸려있었다. 직접 검색했을경우 65만개 정도의 데이터가 나오지만 그 데이터를 전부 분석할 수가 없다는 문제가 생긴 것이다.

    "네이버 검색은 뉴스(news.json), 블로그(blog.json) 및 까페(cafearticle.json)의 요청 End Point 형식만 다르고 전달하는 매개 변수의 형식은 동일하다. 
    query	검색어 파라미터(UTF-8 인코딩 형식)
    start	검색의 시작점(최대값 1,000)
    display	1회 검색에 가지고 올 데이터 레코드 수(최대 100)
    sort	sim(유사도순: 기본값), date(날짜순)

    앞에서 언급하였지만 3월 초 현재 ‘탄핵’이라는 검색어로 조회되는 기사는 43만건이다. 이를 REST API를 이용한다면, 1회 검색할 수 있는 display값이 100 이므로 4,300(430,000 / 100)회의 조회를 해야 한다. 그러나 검색의 시작점 최대가 1,000이므로 우리는 최대 1,099건의 데이터를 가지고 올 수 있다(start=999, display=100으로 전달 하는 경우). 결국 그 이후(시간상은 이전)의 데이터는 검색할 수 있는 방법이 없는 것이다(유료로 계약하면 달라질 것으로 예상한다)."
<br>
출처 = https://ericnjennifer.github.io/python_crawling/2018/01/21/PythonCrawling_Chapt9.html <br>
<br>
하단의 코드가 api를 활용하기 위한 코드였는데 검색결과가 while문을 이용해 반복해보아도 최대 1000건에 불과해 네이버api를 사용할 수가 없었다.

````python
import os
import sys
import urllib.request
client_id = "zy7M6Ej7IaGsr6hR5yzx"# 네이버 api 사용 신청시 지급받는 아이다와 시크릿
client_secret = "F26g01FBkk" 
start_num = 1 # 첫 검색 시작점
encText = urllib.parse.quote("비트코인 | 가상화폐 | 이더리움 | 리플") # 검색할 내용을 utf-8형식으로 바꿔준다.
while start_num < 1102:
    parameter = "&start=" + str(start_num) +"&display=100&sort=date" # 날짜순으로 정렬하고 한번 검색에 100개의 데이터를 보게 설정했다.
    url = "https://openapi.naver.com/v1/search/news?query=" + encText + parameter
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode==200): # 에러 여부 판단
        response_body = response.read()
        print(response_body.decode('utf-8'))
    else:
        print("Error Code:" + rescode)
    start_num += 100
````

````python
import requests
from bs4 import BeautifulSoup

start_num = 1 #페이지 시작 번호
page_num = 10 #끝페이지 번호 이 경우에는 1~10페이지까지 검색하게 된다.

url = "https://search.naver.com/search.naver?&where=news&query=%EA%B0%80%EC%83%81%ED%99%94%ED%8F%90%20%7C%20%EB%B9%84%ED%8A%B8%EC%BD%94%EC%9D%B8%20%7C%20%EC%9D%B4%EB%8D%94%EB%A6%AC%EC%9B%80%20%7C%20%EB%A6%AC%ED%94%8C&sm=tab_pge&sort=1&photo=0&field=0&reporter_article=&pd=3&ds=1997.09.01&de=2017.12.31&docid=&nso=so:dd,p:from19970901to20171231,a:all&mynews=0&start="
date = []
date_num = -1


for start_num in range(start_num, page_num + 1): #1번부터 10번까지의 각 페이지 돌입
    num = start_num * 10 + 1 #검색시 &start= 뒤에 들어갈 숫자는 페이지 수가 아닌 항목 수로 새므로 페이지 수 * 10을 해줘야 한다.
    url_all = url + str(num) 

    header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.102 Safari/537.36'}
    naver = requests.get(url_all, headers = header)
    naver_html = naver.text
    naver_parse= BeautifulSoup(naver_html, 'html.parser')

    
    dates = []
    

    for selector_num in range(num, num + 10): #각 페이지의 10개의 항목에서 뉴스 관련 데이터 추출
        selector = '#sp_nws' + str(selector_num) + ' > dl > dd.txt_inline'
        dates = (str(naver_parse.select(selector)).split()) #날짜가 어느 html 태그에 들어가지 않음으로 우선 공백으로 나눔(이미지 참고)
        num_chk = 0
        while num_chk < len(dates):
            if any(c.isalpha() for c in dates[num_chk]): #리스트로 나눠진 html에서 영어가 있는지의 여부로 영어가 있으면 지워 날짜만을 남김
                del(dates[num_chk])
            else:
                num_chk += 1
            
        dates = dates[0].replace("."," ").strip().split() #2018.01.10 같은 형식의 날짜를 리스트 형태로 쪼개 ['2018','10','02']와 같이 만듬
        

        if date_num == -1: #추출한 데이터 정리 각 날짜에 몇개가 있는가를 ['2018','01','02',3] 뒤에 숫자를 추가함으로써 표현
            dates.append(1)
            date.append(dates)
            date_num = date_num + 1
        else:
            if dates[2] == date[date_num][2]:
                date[date_num][3] += 1
            else:
                dates.append(1)
                date.append(dates)
                date_num = date_num + 1
                
search_date = open("search_date.txt",'w') #데이터 정리하려던 흔적
search_date.write("")
search_date.close()

print(date)
````

import requests
import datetime
from bs4 import BeautifulSoup

url = "https://search.naver.com/search.naver?where=news&query=%EA%B0%80%EC%83%81%ED%99%94%ED%8F%90%20%7C%20%EB%B9%84%ED%8A%B8%EC%BD%94%EC%9D%B8%20%7C%20%EC%9D%B4%EB%8D%94%EB%A6%AC%EC%9B%80%20%7C%20%EB%A6%AC%ED%94%8C&sm=tab_opt&sort=0&photo=0&field=0&reporter_article=&pd=3&ds={0}&de={0}&docid=&nso=so%3Ar%2Cp%3Afrom{1}to{1}%2Ca%3Aall&mynews=0&mson=0&refresh_start=0&related=0"
#초기 검색어의 링크 수정을 원한다면 ?where= 부분에 원하는 카테고리를 query= 부분에 원하는 검색어를 넣어주면 된다.

end_date = "2017.12.31" #검색 종료 날짜

def auto_date_count(start_date):
    '''(str) -> str
    
    add one day to given date. date should be given as "yyyy.mm.dd"
    
    auto_date_count("2018.01.12")
    "2018.01.13"
    '''
    dump = start_date.split(".")
    date = datetime.date(int(dump[0]), int(dump[1]), int(dump[2]))
    return(str(date + datetime.timedelta(days=1)).replace("-","."))

search_date = open("search_date.txt",'w')       

while start_date != auto_date_count(end_date): # 시작 날짜가 종료 날짜와 같을 때 까지 진행함. 종료 날짜는 시작 날짜보다 커야함.
    url_all = url.format(start_date, start_date.replace(".",""))

    naver = requests.get(url_all, headers = header)
    naver_html = naver.text
    naver_parse= BeautifulSoup(naver_html, 'html.parser')
    
    selector = '#main_pack > div.news.mynews.section._prs_nws > div.section_head > div.title_desc.all_my > span' # 검색결과 수의 selector

        dumy_list.append(0)
    else:
        dumy_list.append(int(str(dates).replace("<"," ").replace(">"," ").replace("건","").replace(",","").strip().split()[4])) # 데이터 가공 
    
    data.append(dumy_list)
    start_date = auto_date_count(start_date)
    search_date.write(dumy_list[0] + "\t" + dumy_list[1] + "\t" + dumy_list[2] + "\t" + str(dumy_list[3]) + "\n")


 

search_date.close()
print(data)

### 기사 보도 횟수 데이터 가공

````python
def read_line_search(file_name):
    '''(str)
    
    if you give file name of data then it gives you monthly data back.
    
    '''
    data_file = open(file_name ,'r') #file_name의 파일을 연다.

    name = file_name
    name = name.split(".")
    name = name[0] + "_mon.txt" #file_name을 바탕으로 출력할 파일 이름 결정

    writing_file = open(name, 'w') #출력 파일 오픈

    num = -1 #초기값 설정
    size = []
    volume = 0

    for dump in reversed(data_file.readlines()):#데이터 읽어오기
        dump_list = dump.split() 
        
        if num == -1: #초기 데이터값 주기
            size.append(dump_list[0])
            size.append(dump_list[1])
            volume += int(dump_list[3])
            num += 2
        
        else:  # 기존 데이터가 있다면
            if dump_list[1] == size[1]: #월이 같을 경우 숫자만 추가
                volume += int(dump_list[3])
                num += 1
            
            else: #월이 다를 경우 기존 데이터를 기록하고 새 데이터 추가
                writing_file.write(size[0] + "\t" + size[1] + "\t" + str(round(volume/num)) + "\t" + str(volume) + "\n")
                num = 0
                size.clear()
                volume = 0
                size.append(dump_list[0])
                size.append(dump_list[1])
                volume += int(dump_list[3])
                num += 1
    
    writing_file.write(size[0] + "\t" + size[1] + "\t" + str(round(volume/num)) + "\t" + str(volume) + "\n") #데이터 출력 파일에 기록

    writing_file.close()
    data_file.close()


read_line_search("search_data.txt")
````

### 코인 데이터 가공

````python
def read_line_coin(file_name):
    '''(str)
    
    f you give file name of data then it gives you monthly data back.
    
    '''
    size = []#초기값 설정
    num = -1
    high_price = 0
    volume = 0

    data_file = open(file_name, 'r') #데이터 파일 읽어오기

    name = file_name
    name = name.split(".")
    name = name[0] + "_mon.txt" #데이터 파일 이름을 바탕으로 출력 파일 이름 결정

    writing_file = open(name, 'w') #출력 파일 열기
    
    for dump in data_file.readlines(): #데이터 파일 읽기
        if any(c.isdigit() for c in dump) == True: #데이터 파일의 매 줄마다 영문 포함 여부 판단(노데이터 검출 위해)
            dump_list = dump.replace("\t", " ").replace(",","").replace(".", " ", 3).split()
            if dump_list[7] == "-": #거래량 데이터가 없을 경우
                volume = "NODATA"
                if dump_list[1] != size[1]: #월이 다를 경우 기록
                    writing_file.write(str(size[0]) + "\t" + str(size[1]) + "\t" + str(round(high_price/num, 3)) + "\t" + str(volume)+ "\t" + str(volume) + "\n")
                    size.clear()
                    volume = 0
                    high_price = 0
                    size.append(dump_list[0])
                    size.append(dump_list[1])
                    high_price += float(dump_list[4])
                    num += 1
                else: #같을 경우 숫자만 추가
                    high_price += float(dump_list[4])
                    num += 1
                          
            elif num == -1: #초기 데이터 값 설정 
                size.append(dump_list[0])
                size.append(dump_list[1])
                volume += float(dump_list[7])
                high_price += float(dump_list[4])
                num = 1
            
            elif dump_list[1] != size[1]: #월이 다를 경우 기록
                writing_file.write(str(size[0]) + "\t" + str(size[1]) + "\t" + str(round(high_price/num, 3)) + "\t" + str(round(volume/num)) + "\t" + str(volume) + "\n")
                size.clear()
                volume = 0
                high_price = 0
                size.append(dump_list[0])
                size.append(dump_list[1])
                volume += float(dump_list[7])
                high_price += float(dump_list[4])
                num = 1
            else: #월이 같은 경우 숫자만 추가
                volume += float(dump_list[7])
                high_price += float(dump_list[4])
                num += 1
    #최종 기록
    writing_file.write(str(size[0]) + "\t" + str(size[1]) + "\t" + str(round(high_price/num, 3)) + "\t" + str(volume) + "\t" + str(volume) + "\n")

    data_file.close()
    writing_file.close()

read_line_coin("bit_data.txt")
read_line_coin("eth_data.txt")
read_line_coin("xrp_data.txt")
````

````python
import matplotlib.pyplot as plt

def draw_graph(coin):
    '''(str)
    
    get monthly data and out graph ok good.
    
    '''
    
    coin_file = open(coin + "_data_mon.txt", "r") #데이터 파일 오픈
    search_file = open("search_data_mon.txt", "r") 

    y1 = []
    y2 = []
    date = []

    for search in reversed(search_file.readlines()): #읽어서 값 추가
        search = search.split()
        y1.append(float(search[2]))

    for data in reversed(coin_file.readlines()): 
        data = data.split()
        y2.append(float(data[2]))
        date.append(data[0].replace("20", "") + "." + data[1])

    num = len(y1) - len(y2) #길이 비교

    if num > 0: #날짜가 길 경우 삭제
        a = 0
        while a < num:
            del y1[0]
            a += 1

    fig , ax1 = plt.subplots() #이중축 설정
    ax2 = ax1.twinx()


    ax1.plot(date, y1, color="b", linestyle='--', marker="o", label= "news_amount") #그래프 설정
    ax2.plot(date, y2, color="r", linestyle='--', marker="^", label= "coin_price")
    

    if coin == "bit": #그래프 x축 보이는 것 수정
        ax1.set_xticks(["13.05","13.09","14.01","14.05","14.09","15.01","15.05","15.09","16.01","16.05","16.09","17.01","17.05","17.09","17.12"])
    elif coin == "xrp":
        ax1.set_xticks(["13.08","13.12","14.04","14.08","14.12","15.04","15.08","15.12","16.04","16.08","16.12","17.04","17.08","17.12"])
    elif coin == "eth":
        ax1.set_xticks(["15.08","15.11","16.02","16.05","16.08","16.11","17.02","17.05","17.08","17.11","17.12"])


    ax1.set_xlabel("date") #텍스트 추가
    ax1.set_ylabel("Monthly average news amount(num)")
    plt.ylabel("Monthly average coin price(dollar)")

    plt.title(coin + " coin price & search amount", size = "18")
    plt.rcParams["figure.figsize"] = (14,4) #크기 설정

    ax2.legend()
    ax1.legend(bbox_to_anchor=(0.6, 1.0)) #범례 추가
    plt.show()
    

    search_file.close()
    coin_file.close()

draw_graph("xrp")
draw_graph("eth")
draw_graph("bit")
````
